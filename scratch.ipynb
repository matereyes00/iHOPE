```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the table
columns = ("Precision", "Recall", "F1-Score", "Support")
rows = ["0.0", "1.0", "Accuracy", "Macro Avg", "Weighted Avg"]

# Data in table format
cell_text = [
    [0.56, 0.93, 0.70, 426],
    [0.91, 0.49, 0.64, 606],
    ["", "", 0.67, 1032],
    [0.74, 0.71, 0.67, 1032],
    [0.77, 0.67, 0.66, 1032]
]

# Create a figure and axis
fig, ax = plt.subplots()

# Hide the axes
ax.xaxis.set_visible(False) 
ax.yaxis.set_visible(False)
ax.set_frame_on(False)

# Create the table
table = ax.table(cellText=cell_text, rowLabels=rows, colLabels=columns, cellLoc='center', loc='center')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Add title and additional metrics
plt.title("Classification Report (K = 6)", pad=20)
plt.figtext(0.1, 0.05, "MCC: 0.4470664137233498", horizontalalignment='left', fontsize=12)
plt.figtext(0.1, 0.02, "Overall F1: 0.7013215859030837", horizontalalignment='left', fontsize=12)

# Adjust layout
plt.subplots_adjust(left=0.2, top=0.8)
plt.savefig("classification_report.jpg", format='jpg', dpi=300, bbox_inches='tight')
# Show the plot
plt.show()
```
